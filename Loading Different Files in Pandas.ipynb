# isin() method<br>
a. <b>isin()</b> is predefined method in Series class.<br>
b. We should access this method by using Series object.<br>
c. By using this method we can select data from DataFrame.<br>


```python
#Filtering DataFrame by using loc isin() method
import pandas as pd
df1 = pd.read_csv("sales4.csv")
a = ["Macbook Pro Laptop"]
b = df1.Product_Name.isin(a)
df2 = df1[b]
print(df2)
```

```python
a = ["34in Ultrawide Monitor", "Macbook Pro Laptop"]
b = df1['Product_Name'].isin(a)
df2 = df1[b]
df2
```
